**Question 3:** Write a Python program to check if a number is positive, negative, or zero.

```python
number = input("What's the number?\n")

if int(number) > 0:
    print(f"The number {number} is positive.")
elif int(number) < 0:
    print(f"The number {number} is negative.")
elif int(number) == 0:
    print(f"The number is {number}.")
```

**Question 4:** Write a Python program to find the largest of three numbers.

```python
number1 = input("Enter the first number:\n")
number2 = input("Enter the second number:\n")
number3 = input("Enter the third number:\n")

if int(number1) > int(number2) and int(number1) > int(number3):
    print(f"The largest number is {number1}.")
elif int(number2) > int(number1) and int(number2) > int(number3):
    print(f"The largest number is {number2}.")
else:
    print(f"The largest number is {number3}.")
```

**Question 7:** Write a Python program to swap the values of two variables.

```python
number1 = input("Write the first number:\n")
number2 = input("Write the second number:\n")

def swap_numbers(num1, num2):
    aux = num1
    num1 = num2
    num2 = aux
    return num1, num2

print(f"---Numbers before swapping---\nFirst number -> {number1}\nSecond number -> {number2}\n")
num1, num2 = swap_numbers(number1, number2)
print(f"---Numbers after swapping---\nFirst number -> {num1}\nSecond number -> {num2}\n")
```
