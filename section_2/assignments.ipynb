**Question 7:** Write a Python program to swap the values of two variables.

```python
number1 = input("Write the first number:\n")
number2 = input("Write the second number:\n")

def swap_numbers(num1, num2):
    aux = num1
    num1 = num2
    num2 = aux
    return num1, num2

print(f"---Numbers before swapping---\nFirst number -> {number1}\nSecond number -> {number2}\n")
num1, num2 = swap_numbers(number1, number2)
print(f"---Numbers after swapping---\nFirst number -> {num1}\nSecond number -> {num2}\n")
```

## 3. Basic Operators (Arithmetic, Comparison, Logical)

**Question 11:** Write a Python program to perform arithmetic operations: addition, subtraction, multiplication, and division.

```python
print("---Calculator---\n")
print("[A] Addition")
print("[S] Subtraction")
print("[M] Multiplication")
print("[D] Division\n")
option = input("Choose the operation (letter only):\n")


def addition(*args):
    numbers = args[0].split(", ")
    
    int_list = [int(n) for n in numbers]
        
    return sum(int_list)


def subtraction(*args):
    numbers = args[0].split(", ")
    
    int_list = [int(n) for n in numbers]
    
    result = int_list[0]
    
    for n in int_list:
        if n == int_list[0]:
            continue
        
        result -= n
    
    return result


def multiplication(*args):
    numbers = args[0].split(", ")
    
    int_list = [int(n) for n in numbers]
    
    result = int_list[0]
    
    for n in int_list:
        if n == int_list[0]:
            continue
        
        result *= n
    
    return result


def division(*args):
    numbers = args[0].split(", ")
    
    int_list = [int(n) for n in numbers]
    
    result = int_list[0]
    
    for n in int_list:
        if n == int_list[0]:
            continue
        
        result /= n
    
    return result    


if option == "A" or option == "a":
    numbers = input("Write the numbers to add separated by commas:\n")
    print(f"Result: {addition(numbers)}")
elif option == "S" or option == "s":
    numbers = input("Write the numbers to subtract separated by commas:\n")
    print(f"Result: {subtraction(numbers)}")
elif option == "M" or option == "m":
    numbers = input("Write the numbers to multiply separated by commas:\n")
    print(f"Result: {multiplication(numbers)}")
elif option == "D" or option == "d":
    numbers = input("Write the numbers to divide separated by commas:\n")
    print(f"Result: {division(numbers)}")
    
    
```
